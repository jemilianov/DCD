```python
import numpy as np
import matplotlib.pyplot as plt
import basic_stats as bs
import modelos_lineales as ml
```

```python
#X1,y1,X2,y2=ml.datos_solapados(100)
X1,y1,X2,y2=ml.datos_lin_separables(0.3,100)
X_train,y_train,X_test,y_test=ml.split_train_test(X1,y1,X2,y2,70)    

clf=ml.Perceptron(T=1)
clf2=ml.modelo_basico()

X=ml.centra_datos(X_train)
y=y_train
clf.fit(X,y)

plt.scatter(X[:,0][y>0], X[:,1][y>0],facecolor='orangered', marker='$\\bigoplus$', edgecolor='none',s=90,label='B2')
plt.scatter(X[:,0][y<0], X[:,1][y<0],facecolor='royalblue', marker='$\\ominus$', edgecolor='none',s=90,label='B1')

# Output the values
xx,yy=ml.plot_margin(X[y_train>0], X[y_train<=0], clf)
ymin,ymax=plt.ylim()
plt.axis([xx[0],xx[1],ymin,ymax])
plt.plot(xx, yy, "k")

X_test=ml.centra_datos(X_test)
y_t=clf.predict(X_test)

plt.scatter(X_test[:,0][y_test>0], X_test[:,1][y_test>0],facecolor='g', marker='$\\bigoplus$', edgecolor='none',s=90,label='B2')
plt.scatter(X_test[:,0][y_test<0], X_test[:,1][y_test<0],facecolor='k', marker='$\\ominus$', edgecolor='none',s=90,label='B1')

ex,pr = clf.metricas(y_t,y_test)
print("Exactitud   " + str(ex) + "%")
print("Precision   " + str(pr) + "%")

#Modelo Basico
xx,yy=clf2.fit(X,y)
plt.plot([xx[0],yy[0]],[xx[1],yy[1]],color='olive')

plt.show()
```
